```python
import pandas as pd
data=pd.read_csv("F:/ML/movie_metadata.csv")
df = data.dropna(axis = 0)
print(df.columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.array(df[['actor_1_facebook_likes']])
x2=np.array(df[['actor_2_facebook_likes']])
x3=np.array(df[['actor_3_facebook_likes']])
y=(x1+x2+x3)/3
x=np.array(df[['imdb_score']])
plt.scatter(y,x,color='black')
plt.xlabel('actor_facebood_likes')
plt.ylabel('imdb_score')
plt.show()

x1=np.array(df[['movie_facebook_likes']])
x2=np.array(df[['director_facebook_likes']])
x3=np.array(df[['cast_total_facebook_likes']])
x=(x1+x2+x3)/3
y=np.array(df[['imdb_score']])
plt.scatter(x,y,color='black')
plt.xlabel('other_facebook_likes')
plt.ylabel('imdb_score')
plt.show()
```

  For the convenience, I divided the all the variables about facebook likes into two groups. One group is the mean of all three actors and the other is the mean of other facebook likes such as movie facebook likes and director facebook likes. From the first graph, we can see that high facebook likes do not mean that the movie is high of score rating. From the second scatter plot, we can find movie have more facebook likes tend to have high imdb scores around 7.5. We know that other facebook likes include director facebook likes. So we assume that director facebook likes have more impact on the movie rating.

```python
x=np.array(df[['imdb_score']])
y=np.array(df[['num_critic_for_reviews']])
plt.scatter(y,x,color='black')
plt.ylabel('imdb_score')
plt.xlabel('critic reviews')
plt.show()   

x=np.array(df[['imdb_score']])
y=np.array(df[['num_user_for_reviews']])
plt.scatter(y,x,color='black')
plt.ylabel('imdb_score')
plt.xlabel('user reviews')
plt.show()   

```

We can also see that the difference of two graphs. The first one is critic reviews and the second one is user reviews. Although the number of critic is less than the number of user, the distribution of imdb score of critic reviews is more uniform than the user. In first scatter plot, although reviews of movies below 200 have high socres and low score, with the increasing of critic reviews, the low scores gradually disappear  because the critic is perfer to write review for good movies. In contrst, the most of user reviews are below 2000 which mean that user reviews have less influence on imdb score. 

```python
y=np.array(df[['imdb_score']])
x1=np.array(df[['duration']]) 
plt.scatter(x1,y,color='black')
plt.ylabel('imdb_score')
plt.xlabel('duration')
plt.show()  
```

The duration less than 100 years not only have high scores but also have low socres. But we can see the duration between 100 and 150 years, the scores of movies are gradually increasing.

```python
x=np.array(df[['gross']])
y=np.array(df[['imdb_score']])
plt.scatter(x,y,color='black')
plt.xlabel('gross')
plt.ylabel('imdb_score')
plt.show()   
```

The distuibution of the gross with imdb score is similar with the duration.
